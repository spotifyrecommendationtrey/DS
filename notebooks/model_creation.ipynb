```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial import KDTree
import pickle
```

```python
df = pd.read_csv('../data/merged_data.csv')
```

```python
scaler = MinMaxScaler()
```

```python
df.head()
```

```python
df['track_id'][69]
```

```python
X = df.drop(['artist_name', 'track_id', 'track_name', 'duration_ms', 'popularity'], axis=1)
scaler.fit_transform(X)
```

```python
model = KDTree(X)
```

```python
result = model.query(df[df['track_id'] == '3Rx1zM3nDFQzAOYg9Hd0D4'].drop(['artist_name', 'track_id', 'track_name', 'duration_ms', 'popularity'], axis=1), k=11)
```

```python
df[df['track_id'] == '3Rx1zM3nDFQzAOYg9Hd0D4']
```

```python
result[1][0][1:11]
```

```python
# This is a custom song with the audio features values below
result = model.query([[0.01, 0.5, 0.5, 0.1, 1, 0.1, -5, 1, 0.5, 190, 4, 0.5]], k=11)
```

```python
result[1][0][1:]
```

```python
# See that the suggested IDs correspond to the IDs in the database
df.iloc[result[1][0][1:]]
```

```python
pickle.dump(model, open('suggestion_model_2.pkl', 'wb'))
```
